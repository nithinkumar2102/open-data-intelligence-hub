```python
year_cols = df.columns[4:]

df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')

avg_urbanization = df[year_cols].mean()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(year_cols, avg_urbanization, marker='o')
plt.xticks(rotation=45)
plt.title("Average Urbanization Trend")
plt.xlabel("Year")
plt.ylabel("Urban Population (%)")
plt.grid(True)
plt.show()
```

```python
latest_year = '2025 [YR2025]'

top5 = df[['Country Name', latest_year]].sort_values(
    by=latest_year,
    ascending=False
).head(5)

plt.figure(figsize=(8,5))
plt.bar(top5['Country Name'], top5[latest_year])
plt.title('Top 5 Countries by Urban Population % (2025)')
plt.xlabel('Country')
plt.ylabel('Urban Population (%)')
plt.xticks(rotation=45)
plt.show()
```
