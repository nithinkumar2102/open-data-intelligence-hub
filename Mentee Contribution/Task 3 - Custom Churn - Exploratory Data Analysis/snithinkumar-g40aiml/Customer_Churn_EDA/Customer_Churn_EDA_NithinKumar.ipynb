```python
years = ['2001 [YR2001]','2002 [YR2002]','2003 [YR2003]','2004 [YR2004]','2005 [YR2005]']

afghanistan = df[df['Country Name'] == 'Afghanistan']

trend = afghanistan[years].T

trend.plot(figsize=(8,5))
plt.title("Urban Population Trend - Afghanistan")
plt.xlabel("Year")
plt.ylabel("Urban Population %")
plt.show()
```

```python
top_2019 = df[['Country Name','2019 [YR2019]']]

top_2019 = top_2019.sort_values('2019 [YR2019]', ascending=False)

top_2019.head(10)
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    x='2019 [YR2019]',
    y='Country Name',
    data=top_2019.head(10)
)

plt.title("Top 10 Countries by Urban Population % (2019)")
plt.show()
```
